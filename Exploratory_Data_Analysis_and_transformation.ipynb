# Exploratory data analysis and transformation

This notebook is for exploring the data, getting a sense of how it is structured and, and making nessecary transformations

```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv('data/race_results.csv')
country_key=pd.read_csv('data/nationalities_key.csv')
```

rider nationalites need to be matched up wiht country names based on the provdied key

```python
d['rider_country']=d['rider_country'].astype('category')
country_key['rider_country']=country_key['rider_country'].astype('category')
d=d.merge(country_key,on="rider_country",how='left')
d['race_country']=d['race_country'].str.strip()
```

### determining if a rider was racing in their home country

```python
d['home_race']=(d['race_country']==d['rider_country_new'])*1
d['home_race'].sum()
```

```python
d.info()
```

Converting race times to be viewed as actual times

```python
bad_results=['DNF','DNQ','DSQ','DNS','',np.nan]
def convert_time_to_decimal(race_time):
    if race_time in bad_results:
        return(np.nan)
    else:
        mins, sec =race_time.split(':')
        return(int(mins)+(float(sec)/60))
    
d['run_time']=d['Run 1'].apply(convert_time_to_decimal)
```

Quick look at the distribution of race times - they look a bit right skewed but less that I expected

```python
sns.histplot(data=d,x='run_time')
```

## recoding races times as percent back from winning

This should help normlaize differences among tacks

```python
d['race_id']=d['track_name']+d['Date']
races=d['race_id'].unique()
d['percent_time_back']=""

for i in range(len(races)):
    try:
        d.loc[(d['race_id']==races[i]) & (d['sex']=='Men'),'percent_time_back']= (d.loc[(d['race_id']==races[i]) & (d['sex']=='Men'),'run_time']/min(d.loc[(d['race_id']==races[i]) & (d['sex']=='Men'),'run_time'])-1)*100
        d.loc[(d['race_id']==races[i]) & (d['sex']=='Women'),'percent_time_back']= (d.loc[(d['race_id']==races[i]) & (d['sex']=='Women'),'run_time']/min(d.loc[(d['race_id']==races[i]) & (d['sex']=='Women'),'run_time'])-1)*100
    except:
        print('trouble with',races[i])
    
```

lets see what these data look like

```python
sns.histplot(data=d,x='percent_time_back')
```

## coding new columns for the best and worst 25% of a riders results

```python
riders=d['Name'].unique()
print('before trimming ',len(riders))

#subsetting riders with 20 or more results
num_races = d.groupby(['Name'])['Name'].count()
modeling_names=num_races[num_races>19].index
#425 riders have done 20 or more races
d=d[d['Name'].isin(list(modeling_names))]
riders=d['Name'].unique()
print('after trimming ',len(riders))
```

```python
d['best_results']=0
d['worst_results']=0
d['percent_time_back']=pd.to_numeric(d['percent_time_back'])

for i in range(len(riders)):
    try:
        times=d.loc[d['Name']==riders[i],'percent_time_back'].dropna()
                      
        worst_cutoff=np.quantile(times,0.75)
        best_cutoff=np.quantile(times,0.25)
        max_time=times.max()
        d.loc[d['Name']==riders[i],'percent_time_back']=d.loc[d['Name']==riders[i],'percent_time_back'].fillna(max_time)

        d.loc[(d['Name']==riders[i]) & (d['percent_time_back']<=best_cutoff),'best_results']=1
        d.loc[(d['Name']==riders[i]) & (d['percent_time_back']>=worst_cutoff),'worst_results']=1
    except:
        print(riders[i])

## Ondrej BARTA never qualified for a race so the code did
```
